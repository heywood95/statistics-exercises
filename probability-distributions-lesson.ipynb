### Terminology

#### Random variable

- Represents the outcome of a random event

#### Probability

- Likelihood of an event
- A value from [0,1}

### Probability distribution

- a mathematical function
- describe probability of an outcome
- Can be displayed as a graph


### Discrete versus Continuous Variables

- Discete
    - Number of customer service tickets
    - Number of students in a class
    
    
- Continuous
    - Height
    - Temperature
    - Salaries
    

## Distrubtions with SciPy

### 4 Distributions

#### Uniform distribution

- Equal likelihood of each outcome

#### Binomial Distribution

- Exactly two outcomes (success or failure)
- P(success) = 1 - P(failure)


### Normal Distribution

- Models a *continuous* random variable

- Defined using: mean and standard deviation
    
    
#### Poisson Distribution

- Discrete random variables
- Model the liklihood of a certain number of events occuring in a fixed length of time

    Example: A restaurant serving 20 customers in an hour





### Process

1. Detrmine what we are calculation

- Random variable = given value
- Random variable <= given value
- Random variable >= given value

2. Determine what you have and what you need to find

Find the probability for a given value

E.g. Likelihood of having 12 customers

Find the value associated with a probability

E.g. What GPA correspond to the 10% of a class

### Functions

#### Probability Mass Function (PMF)
-Distribution of a *discrete* random variable

### Probability Density Function
- Distribution of a *continuous* random variable

### Cumulative Distribution
- Probability of an outcome OR all the outcomes below it

    Example: 
        - Value of 3 or lower
        - At most 3

### Percent Point Function (PPF)

- AKA Quantile 
- Return a value of the random variable associated with a given probability

### Survival Function

- Probability of a random variable being GREATER than a given value


### Inverse Function

- Used in the same scenario as survival function
- Find the value asscoiated with a probability




```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import viz # curriculum viz example code

np.random.seed(123)
```

```python
# rvs - rand value generation

# Generate random values to reflect dice rolls

die_distribution = stats.randint(1, 7)

```

```python
# no values entered
die_distribution.rvs()
```

```python
# Specify the size of an array
die_distribution.rvs(5)
```

```python
# For a matrix
# Use the dimension (row, col)

die_distribution.rvs((5, 5))
```

```python
# Set number of trials
n = 10000


x = die_distribution.rvs(n)


n = 10_000

x = die_distribution.rvs(n)

plt.hist(x, bins=range(0, 9), align='left', width=1, edgecolor='black')
plt.title(f'Outcome of {n:,} Dice Rolls')

```

```python
# Use a PMF to find the probability that a roll = 3

die_distribution.pmf(3)
```

```python
# Use a CDF for the prbability that a roll is 3 or lower

die_distribution.cdf(3)
```

```python
## Use a PPF to find the value associated with 5/6
## 5/6
die_distribution.ppf(5/6)
```

```python
# Mean
m = 12000

# Standard deviation
s = 2000

# Defining the normal distribution
sales = stats.norm(m, s)

# Using the PPF to find the cutoff
# Where 90% of the data are less than the cutoff

top_10_percent_cutoff = sales.ppf(.9)


print('${:,.2f}'.format(top_10_percent_cutoff))
```

```python
viz.distributions_example7(m, s)
```

```python
# Find the probability < 10,000

# Cumulative

p = sales.cdf(10_000)

print(f'Ony any given day, there\'s a {p:.1%} chance we sell less than $10,000.')
```
